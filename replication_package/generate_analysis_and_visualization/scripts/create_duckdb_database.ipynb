```python
%%sql
CREATE TABLE new_snapshot(
    Project VARCHAR, Owner VARCHAR, Manager VARCHAR, Category VARCHAR, Key VARCHAR, Priority VARCHAR, Status VARCHAR,
    Reporter VARCHAR, Assignee VARCHAR, Components VARCHAR, SummaryTopWords VARCHAR, DescriptionTopWords VARCHAR, 
    CommentsTopWords VARCHAR, CreationDate TIMESTAMP, ResolutionDate TIMESTAMP, AffectsVersions VARCHAR, 
    FixVersions VARCHAR, NoComments INTEGER, FirstCommentDate TIMESTAMP, LastCommentDate TIMESTAMP, NoWatchers INTEGER,
    NoAttachments INTEGER, FirstAttachmentDate TIMESTAMP, LastAttachmentDate TIMESTAMP, NoAttachedPatches INTEGER,
    FirstAttachedPatchDate TIMESTAMP, LastAttachedPatchDate TIMESTAMP, InwardIssueLinks VARCHAR, 
    OutwardIssueLinks VARCHAR, HasMergeCommit INTEGER, CommitsMessagesTopWords VARCHAR, NoCommits INTEGER, NoAuthors INTEGER,
    NoCommitters INTEGER, AuthorsFirstCommitDate TIMESTAMP, AuthorsLastCommitDate TIMESTAMP, 
    CommittersFirstCommitDate TIMESTAMP, CommittersLastCommitDate TIMESTAMP, NonSrcAddFiles INTEGER, NonSrcDelFiles INTEGER,
    NonSrcModFiles INTEGER, NonSrcAddLines INTEGER, NonSrcDelLines INTEGER, SrcAddFiles INTEGER, SrcDelFiles INTEGER,
    SrcModFiles INTEGER, SrcAddLines INTEGER, SrcDelLines INTEGER, TestAddFiles INTEGER, TestDelFiles INTEGER, 
    TestModFiles INTEGER, TestAddLines INTEGER, TestDelLines INTEGER, summary_token_number INTEGER, description_token_number INTEGER, 
    summary_char_number INTEGER, description_char_number INTEGER);

COPY new_snapshot FROM '../datasets/snapshot_voting.csv' (AUTO_DETECT TRUE);
```
