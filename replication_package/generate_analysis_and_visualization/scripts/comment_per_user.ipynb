```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
CONTRIBUTORS_BOTS = ["id159", "id733", "id1122", "id1661", "id1706", "id1745"]
INCLUDE_BOTS = False
```

```python
df = pd.read_csv('../datasets/commentslog_voting.csv', sep=';')
```

```python
df
```

```python
# removing bots
if not INCLUDE_BOTS:
    df = df[~df['Author'].isin(CONTRIBUTORS_BOTS)] 
df
```

```python
# get statistics of authors per bug
counts = df.groupby(['Key', 'Author']).size().reset_index(name='Comments')
total_counts = counts.groupby('Key')['Comments'].transform('sum')
counts['Percentage'] = counts['Comments'].div(total_counts).mul(100)
counts['Percentage'] = counts['Percentage'].round(2)
new_df_author_count_by_project = counts[['Author', 'Key', 'Comments', 'Percentage']]

# save dataset
datasetname = '../datasets/comment_author_groupby_bug.csv'
if INCLUDE_BOTS:
  datasetname = '../datasets/comment_author_groupby_bug_bots.csv'
new_df_author_count_by_project.to_csv(datasetname, sep=';')

# preview of dataset where key = 'HADOOP-100003'
new_df_author_count_by_project[new_df_author_count_by_project['Key'] == 'HADOOP-10003']
```

```python
# get general statistics
counts = df['Author'].value_counts().reset_index()
counts.columns = ['Author', 'Comments']
total_counts = counts['Comments'].sum()
counts['Percentage'] = counts['Comments'].div(total_counts).mul(100)
# round to 2 decimal places
counts['Percentage'] = counts['Percentage'].round(2)
new_df_author_count = counts[['Author', 'Comments', 'Percentage']]

# save dataset
authordatasetname = '../datasets/comment_author.csv'
if INCLUDE_BOTS:
  authordatasetname = '../datasets/comment_author_bots.csv'
new_df_author_count.to_csv(datasetname, sep=';')

new_df_author_count
```

```python
# check if the percentage of previous dataset is equal to 100%
print('Sum of percentage: ', new_df_author_count['Percentage'].sum())

# check if all authors are in the dataset
print('Number of unique Authors: ', df['Author'].nunique())
print('Length of dataframe: ', len(new_df_author_count))
```

```python
# import summary dataset
if not INCLUDE_BOTS:
    df_summary = pd.read_csv('../datasets/contributors_summary_dataset_full.csv', sep=';')
else:
    df_summary = pd.read_csv('../datasets/contributors_summary_dataset_full_bots.csv', sep=';')
```

```python
# Rename column 'Contributor' to 'Author'
df_summary = df_summary.rename(columns={'Contributor': 'Author'})

# print all lines with Author = 'id1122' and key = 'HADOOP-10003'
# df_summary[(df_summary['Author'] == 'id1122') & (df_summary['Key'] == 'HADOOP-10003')]

# merge datasets (summary and comments)

##############
# IMPORTANT! #
##############
# the datasets have missing data relative to each other (contributors_summary_dataset_full and comment_author_groupby_bug or new_comment_file)
# Since in this experiment the comments are more important than whether the user was a reporter or assignee, the merge is being done based on the 
# comment dataframe (left join). So we have some NaN data in the dataframe only in the added columns.
df_merge = pd.merge(new_df_author_count_by_project, df_summary[['Author', 'Key', 'Reporter', 'Assignee', 'PMC', 'Official', 'Organization']], on=['Author', 'Key'], how='left')
df_merge
```

```python
# save dataset
df_merge.to_csv(datasetname, sep=';')
```
